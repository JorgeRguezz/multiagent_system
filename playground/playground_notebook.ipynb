```python
import json
from pathlib import Path


def load_json_array(path1: Path, path2: Path, output: Path):
    with open(path1, "r", encoding="utf-8") as f:
        data1 = json.load(f)

    with open(path2, "r", encoding="utf-8") as f:
        data2 = json.load(f)

    merged = data1 + data2

    with open(output, "w", encoding="utf-8") as f:
        json.dump(merged, f, ensure_ascii=False, indent=4)

    print(f"Loaded {len(data1)} objects from {path1}")
    print(f"Loaded {len(data2)} objects from {path2}")
    print(f"Done. Wrote {len(merged)} objects to {output}")
```

```python
path1 = "/home/gatv-projects/Desktop/project/knowledge_inference/report_eval_test.json"
path2 = "/home/gatv-projects/Desktop/project/knowledge_inference/report_eval_test_v2.json"
output_path = "/home/gatv-projects/Desktop/project/knowledge_inference/merged_report_eval_test.json"
load_json_array(path1, path2, output_path)
```
