```python
from PIL import Image
import os

TARGET_SIZE = 518

def resize_with_padding(input_path, output_path, fill_color=(0, 0, 0)):
    """
    Resize an image to 224x224 for DINOv2:
    - Keep aspect ratio
    - Use bicubic interpolation
    - Pad to 224x224 with a solid background (default black)
    """
    img = Image.open(input_path).convert("RGB")

    # Original size
    w, h = img.size

    # Compute scale factor so that the longest side == TARGET_SIZE
    scale = TARGET_SIZE / max(w, h)
    new_w = int(w * scale)
    new_h = int(h * scale)

    # Resize with bicubic interpolation
    img_resized = img.resize((new_w, new_h), resample=Image.BICUBIC)

    # Create new 224x224 canvas and paste the resized image centered
    new_img = Image.new("RGB", (TARGET_SIZE, TARGET_SIZE), fill_color)
    left = (TARGET_SIZE - new_w) // 2
    top = (TARGET_SIZE - new_h) // 2
    new_img.paste(img_resized, (left, top))

    # Ensure output directory exists
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    # Save output (PNG is usually best for game UI crops)
    new_img.save(output_path, format="PNG")

    print(f"Saved resized image to: {output_path}")


if __name__ == "__main__":
    # Example usage:
    # Put your tiny LoL crop path here
    input_image = "/home/gatv-projects/Desktop/project/playground/sam3_testing/cutout_Vladimir_x4_lanczos.png"
    output_image = "/home/gatv-projects/Desktop/project/playground/sam3_testing/resize_testing/cutout_Vladimir_x4_lanczos_resized.png"

    resize_with_padding(input_image, output_image)
 
```
